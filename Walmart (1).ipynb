```python
from scipy.stats import ttest_ind
# Filter the data by customer type
ratings_member = df[df['Customer type'] == 'Member']['Rating']
ratings_normal = df[df['Customer type'] == 'Normal']['Rating']

# Perform the two-sample t-test
t_stat, p_value = ttest_ind(ratings_member, ratings_normal, equal_var=False)  # Assuming unequal variances

# Print the results
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in ratings between Member and Normal customers.")
else:
    print("Fail to reject the null hypothesis: No significant difference in ratings between Member and Normal customers.")
```
